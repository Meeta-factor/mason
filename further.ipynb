```python
def bernoulli_numbers(N):
    B = [sp.Integer(0)] * (N + 1)
    B[0] = sp.Integer(1)

    for m in range(1, N + 1):
        s = sum(sp.binomial(m + 1, k) * B[k] for k in range(m))
        B[m] = -sp.simplify(s / sp.Integer(m + 1)) # type: ignore
    return B

print(bernoulli_numbers(10))
```

```python

def euler_maclaurin_sum(fx, x, a, b, order=3):
    expr = sp.integrate(fx, (x, a, b)) + (fx.subs(x, a) + fx.subs(x, b))/2
    for m in range(1, order + 1):
        B = sp.bernoulli(2*m)
        term = B / sp.factorial(2*m) * ( # type: ignore
            sp.diff(fx, x, 2*m - 1).subs(x, b)
            - sp.diff(fx, x, 2*m - 1).subs(x, a)
        ) # type: ignore
        expr += term
    return sp.collect(sp.expand(expr), [
        fx.subs(x, a), fx.subs(x, b),
        sp.diff(fx, x).subs(x, a), sp.diff(fx, x).subs(x, b)
    ])
```

```python
x, n = sp.symbols('x n', integer=True, positive=True)
k = sp.symbols('k', integer=True)
f = sp.Function('f')
```

```python

S_exact = sp.Sum(f(k), (k, 1, n))
S_em = euler_maclaurin_sum(f(x), x, 1, n)
S_em
```
