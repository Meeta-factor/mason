```python
import sympy as sp
from mason.solver import MIMOSFGSolver,ShannonHappSolver

G11, G12, G21, G22 = sp.symbols("G11 G12 G21 G22")

data_mimo = {
    "edges": [
        ("R1", "C1", G11),
        ("R2", "C1", G12),
        ("R1", "C2", G21),
        ("R2", "C2", G22),
    ],
    "sources": ["R1", "R2"],
    "sinks": ["C1", "C2"],
}

mimo_solver = MIMOSFGSolver(entry_solver_cls=ShannonHappSolver)
mimo_solver.load_from_dict(data_mimo)

G, info = mimo_solver.solve()
print("Transfer matrix:")
G
```

```python
show_result(info)
```
