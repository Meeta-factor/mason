```python
import sympy as sp
import control as ct
import pandas as pd
from mason.visualize import *
import matplotlib.pyplot as plt
from mason.solver import MasonSolver
from mason.adapters.control import *
s = sp.symbols("s")
df = pd.read_csv("../data/loop3.csv")
solver = MasonSolver()
```

```python
data = {
    "edges": [
        (row.start, row.end, sp.sympify(row.gain))
        for _, row in df.iterrows()
    ],
    "source": ["R"],
    "sink": ["Y"],
}
```

```python
solver.load_from_dict(data)
```

```python
G,info = solver.transfer_function("R","Y")
```
