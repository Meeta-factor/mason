```python
from mason import MasonSolver,MIMOSFGSolver
from mason.visualize import *
import pandas as pd
import sympy as sp
df = pd.read_csv("../mason/tikz/load.csv")
solver = MIMOSFGSolver(entry_solver_cls=MasonSolver)
```

```python
data = {
    "edges": [
        (row.start, row.end, sp.sympify(row.gain))
        for _, row in df.iterrows()
    ],
    "sources": ["PL1", "PL2","Pref1","Pref2"],
    "sinks": ["w1", "w2","P12","ACE1", "ACE2"],
}
```

```python
solver.load_from_dict(data)
```

```python
G11,info11 = solver.transfer_function("PL1","w1")
```
