```python
from sympy import symbols, summation, Integer

k, n = symbols('k n')

```

```python
times = 30
S = [None] * times
S[0] = Integer(1)   # type: ignore
for m in range(1, times):
    S[m] = summation(S[m-1].subs(n, k), (k, 1, n)) # type: ignore

for i, expr in enumerate(S):
    print(f"S{i}(n) =", expr)
```

```python
from sympy import Poly, Matrix

# 打印成“矩阵形式”
for i, expr in enumerate(S[1:20], start=1):
    p = Poly(expr, n)
    coeffs = p.all_coeffs()   # 高次 → 低次

    print(f"S{i}: ", coeffs)
```

```python
max_deg = 10

matrix = []

for expr in S[1:10]:
    p = Poly(expr, n)
    coeffs = p.all_coeffs()

    # 左侧补0，让长度一致
    coeffs = [0]*(max_deg - len(coeffs)) + coeffs
    matrix.append(coeffs)

# 打印
for row in matrix:
    print(row)
```

```python
import csv

with open("coeff_matrix.csv", "w", newline="") as f:
    writer = csv.writer(f)
    for row in matrix:
        writer.writerow([str(c) for c in row])
```
