```python
import sympy as sp
```

```python
class Operator:
    def __init__(self, func):
        self.func = func

    def __call__(self, f):
        return self.func(f)

    def __mul__(self, other):
        return Operator(lambda f: self.func(other.func(f)))
```

```python
class SigmaOperator:
    def __init__(self, var, lower=0):
        self.var = var
        self.lower = lower

    def __call__(self, f, m=1):
        f = sp.sympify(f)
        expr = f
        for _ in range(m):
            k = sp.Dummy('k')
            expr = sp.summation(expr.subs(self.var, k), (k, self.lower, self.var - 1))
        return sp.simplify(expr)
```

```python
class DeltaOperator:
    def __init__(self, var):
        self.var = var
        self.E = ShiftOperator(var)

    def __call__(self, f, m=1):
        f = sp.sympify(f)
        expr = f
        for _ in range(m):
            expr = self.E(expr) - expr
        return sp.simplify(expr)
```

```python
class DOperator:
    def __init__(self, var):
        self.var = var

    def __call__(self, f, m=1):
        f = sp.sympify(f)
        return sp.diff(f, self.var, m)
```

```python
class IntegralOperator:
    def __init__(self, var):
        self.var = var

    def __call__(self, f, m=1):
        f = sp.sympify(f)
        expr = f
        for _ in range(m):
            expr = sp.integrate(expr, self.var)
        return sp.simplify(expr)
```

```python
class ShiftOperator:
    def __init__(self, var):
        self.var = var

    def __call__(self, f, k=1):
        return f.subs(self.var, self.var + k)
```

```python
n, x = sp.symbols('n x')

E = ShiftOperator(n)
Delta = DeltaOperator(n)
Sigma = SigmaOperator(n)

D = DOperator(x)
I = IntegralOperator(x)

f = n**2

E(f)        # f(n+1)
```

```python
Delta(f)    # Δf
```

```python
Sigma(f)    # Σf
```
